```python
#probar el principio de pareto

def pareto_principle(column):
    sorted_df = df.sort_values(by=column, ascending=False)
    cumulative_sum = sorted_df[column].cumsum()
    total_sum = df[column].sum()
    sorted_df['cumulative_percentage'] = cumulative_sum / total_sum * 100
    top_20_percentage = sorted_df.head(int(0.2 * len(sorted_df)))['cumulative_percentage'].max()
    return top_20_percentage

pareto_results = {column: pareto_principle(column) for column in df.columns if column not in ['Channel', 'Region']}
pareto_results
```

En general, el 20% superior de los clientes representa más del 50% del gasto total en todas las categorías, cumpliendo con el principio de Pareto.

# Reto 2 - Limpieza y transformación de datos

Si tu conclusión del reto anterior es que los datos necesitan limpieza/transformación, hazlo en las celdas de abajo. Sin embargo, si su conclusión es que los datos no necesitan ser limpiados o transformados, no dudes en saltarte este reto. Si optas por esta última opción, explica los motivos.

```python
#si que hay que hacer algo de limpia y transformacion, porque hay muchas columnas con valores atipicos y con alta colinealidad
```

```python
#eliminar valores atípicos usando el IQR
def eliminar_outliers(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    df_sin_outliers = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
    return df_sin_outliers

# Eliminar valores atípicos del DataFrame
df_limpio = eliminar_outliers(df)

# Verificar el tamaño del DataFrame antes y después de la limpieza
tamaño_antes = df.shape
tamaño_despues = df_limpio.shape

print(tamaño_antes, tamaño_despues)
```

**Tus observaciones aquí**

+ Se utilizaron técnicas de rango intercuartil (IQR) para detectar y eliminar valores atípicos.
+ El tamaño del DataFrame se redujo significativamente después de eliminar los valores atípicos, lo que indica la presencia de numerosos valores extremos en las columnas de gasto.


# Reto 3 - Preprocesamiento de datos

Uno de los problemas del conjunto de datos es que los rangos de valores son notablemente diferentes en las distintas categorías (por ejemplo, `Fresh` y `Grocery` en comparación con `Detergents_Paper` y `Delicassen`). Si hiciste esta observación en el primer reto, ¡has hecho un gran trabajo! Esto significa que no sólo has completado las preguntas de bonificación en el anterior laboratorio de Aprendizaje Supervisado, sino que también has investigado en profundidad sobre [*feature scaling*](https://en.wikipedia.org/wiki/Feature_scaling). ¡Sigue trabajando así de bien!

Diversos rangos de valores en diferentes características podrían causar problemas en nuestra agrupación. La forma de reducir el problema es mediante el escalado de características. Volveremos a utilizar esta técnica con este conjunto de datos.

#### Utilizaremos el `StandardScaler` de `sklearn.preprocessing` y escalaremos nuestros datos. Lee más sobre `StandardScaler` [aquí](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler).

*Después de escalar tus datos, asigna los datos transformados a una nueva variable `customers_scale`.

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler


# Supongamos que df_limpio es tu DataFrame con los datos originales
# Aquí seleccionamos las características de interés
features = ['Detergents_Paper', 'Grocery', 'Milk']
x = df_limpio[features]

# Escalar las características
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

# Aplicar PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(x_scaled)

# Crear un DataFrame con los componentes principales
df_pca = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])

# Añadir las columnas PCA al DataFrame original (sin valores atípicos)
df_limpio.reset_index(drop=True, inplace=True)
df_pca_limpio = pd.concat([df_limpio, df_pca], axis=1)

# Mostrar las primeras filas del DataFrame con PCA
print(df_pca_limpio.head())


```

```python
# Supongamos que df_limpio es tu DataFrame con los datos originales
# Aquí seleccionamos las características de interés
features = ['Detergents_Paper', 'Grocery', 'Milk']
x = df_limpio[features]

# Crear una instancia de StandardScaler
scaler = StandardScaler()

# Ajustar y transformar las características
customers_scale = scaler.fit_transform(x)

# Convertir la matriz escalada a un DataFrame
customers_scale_df = pd.DataFrame(customers_scale, columns=features)

# Verificar las primeras filas del DataFrame escalado
print(customers_scale_df.head())
```

```python

# Visualización de los componentes principales
plt.figure(figsize=(10, 6))
plt.scatter(df_pca_limpio['PC1'], df_pca_limpio['PC2'])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Visualización de los Componentes Principales')
plt.show()
```

# Reto 4 - Agrupación de datos con K-Means

Ahora vamos a agrupar los datos con K-Means primero. Inicia el modelo K-Means, luego ajusta tus datos escalados. En los datos devueltos por el método `.fit`, hay un atributo llamado `labels_` que es el número de cluster asignado a cada registro de datos. Lo que puede hacer es asignar estas etiquetas de nuevo a `customers` en una nueva columna llamada `customers['labels']`. Entonces verá los resultados de cluster de los datos originales.

# Reto 5 - Clustering de datos con DBSCAN

Ahora vamos a agrupar los datos utilizando DBSCAN. Utiliza `DBSCAN(eps=0.5)` para iniciar el modelo y, a continuación, ajusta los datos escalados. En los datos devueltos por el método `.fit`, asigna las `labels_` de nuevo a `customers['labels_DBSCAN']`. Ahora tus datos originales tienen dos etiquetas, una de K-Means y la otra de DBSCAN.

```python
from sklearn.cluster import DBSCAN

# Supongamos que df_limpio es tu DataFrame con los datos originales
# Selecciona las características de interés
features = ['Detergents_Paper', 'Grocery', 'Milk']
x = df_limpio[features]

# Escalar las características
scaler = StandardScaler()
customers_scale = scaler.fit_transform(x)

# Inicializar y ajustar el modelo DBSCAN
dbscan = DBSCAN(eps=0.5)
dbscan.fit(customers_scale)

# Obtener las etiquetas de clúster de DBSCAN
labels_dbscan = dbscan.labels_

# Asignar las etiquetas al DataFrame original
df_limpio['labels_DBSCAN'] = labels_dbscan

# Mostrar las primeras filas del DataFrame con las etiquetas de clúster DBSCAN
print(df_limpio.head())
```

Cuenta los valores en `labels_DBSCAN`.

```python
dbscan_counts = df_limpio['labels_DBSCAN'].value_counts()
print(dbscan_counts)
```

# Bonus Challenge 2 - Cambiar el número de clusters de K-Means

Como hemos mencionado antes, no tenemos que preocuparnos por el número de clusters con DBSCAN porque lo decide automáticamente en función de los parámetros que le enviemos. Pero con K-Means, tenemos que suministrar el parámetro `n_clusters` (si no se suministra `n_clusters`, el algoritmo utilizará `8` por defecto). Debe saber que el número óptimo de clusters varía en función del conjunto de datos. K-Means puede funcionar mal si se utiliza un número incorrecto de clusters.

En el aprendizaje automático avanzado, los científicos de datos prueban diferentes números de clusters y evalúan los resultados con medidas estadísticas (leer [aquí](https://en.wikipedia.org/wiki/Cluster_analysis#External_evaluation)). Hoy no vamos a utilizar medidas estadísticas, sino nuestros ojos. En las celdas de abajo, experimenta con distintos números de conglomerados y visualízalos con gráficos de dispersión. ¿Qué número de clusters parece funcionar mejor para K-Means?

```python
# Your code hereimport matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Definir una lista de números de clústeres para probar
n_clusters_list = [2, 3, 4, 5]

# Configurar el tamaño de la figura para los gráficos
fig, axs = plt.subplots(len(n_clusters_list), 3, figsize=(18, 5 * len(n_clusters_list)))

# Para cada número de clústeres en la lista, aplicar K-Means y visualizar
for i, n_clusters in enumerate(n_clusters_list):
    # Aplicar K-Means
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(customers_scale)
    
    # Obtener etiquetas de clústeres
    labels = kmeans.labels_
    
    # Graficar cada par de características
    # Detergents_Paper vs Milk
    axs[i, 0].scatter(df_limpio['Detergents_Paper'], df_limpio['Milk'], c=labels, cmap='viridis', marker='o', edgecolor='k')
    axs[i, 0].set_title(f'K-Means with {n_clusters} clusters\nDetergents_Paper vs Milk')
    axs[i, 0].set_xlabel('Detergents_Paper')
    axs[i, 0].set_ylabel('Milk')

    # Grocery vs Fresh
    axs[i, 1].scatter(df_limpio['Grocery'], df_limpio['Fresh'], c=labels, cmap='viridis', marker='o', edgecolor='k')
    axs[i, 1].set_title(f'K-Means with {n_clusters} clusters\nGrocery vs Fresh')
    axs[i, 1].set_xlabel('Grocery')
    axs[i, 1].set_ylabel('Fresh')

    # Frozen vs Delicassen
    axs[i, 2].scatter(df_limpio['Frozen'], df_limpio['Delicassen'], c=labels, cmap='viridis', marker='o', edgecolor='k')
    axs[i, 2].set_title(f'K-Means with {n_clusters} clusters\nFrozen vs Delicassen')
    axs[i, 2].set_xlabel('Frozen')
    axs[i, 2].set_ylabel('Delicassen')

plt.tight_layout()
plt.show()

```

**Tus observaciones aquí**

* Viendo los gráficos del k-means yo diria que 3 es la mejor opcion, al menos para detergents_papel vs milk y para grocery vs fresh. Incluso llegaria hasta 4 en detergens vs milk pero preferiria tener más datos. 5 parecen demasiados en todas las opciones, no se distinguen grupos claros
* El problema es con frozen vs delicatessen, no soy capaz de ver unos grupos muy muy claros, como mucho en 2 podriamos intentar agrupar pero hay mucho punto en el cluster del otro.
